```python
import pandas as pd
import numpy as np
import random
from pprint import pprint
from IPython.display import clear_output 

data = pd.read_csv("/content/data.csv")
data
```

```python
news = data[data['relevance'] == 1][['id', 'team1', 'team2', 'first_part_content', 'last_part_content', 'part_body']]
news
```

```python
news['result_match'] = 0
```

```python
good_news = []
bad_news = []
indexes = news.index.tolist()
random.shuffle(indexes)

for index in indexes:
  print('len(good_news)=', len(good_news))
  print('len(bad_news)=', len(bad_news))
  if len(good_news) >= 500 and len(bad_news) >= 500:
    break
  row = news.loc[index]
  print(row['team1'], 'VS', row['team2'])
  pprint(row['first_part_content'], width=150)
  print('--------------------')
  pprint(row['last_part_content'], width=150)
  print('--------------------')
  pprint(row['part_body'], width=150)
  label = int(input())
  if label == 1:
    good_news.append(index)
  else:
    bad_news.append(index)
  clear_output(True)
```

```python
good_news
```

```python
bad_news
```

```python
good_train = good_news[:250]
good_eval = good_news[250:500]

bad_train = bad_news[:250]
bad_eval = bad_news[250:500]
```

```python
train = pd.concat([data.loc[good_train], data.loc[bad_train]])
train
```

```python
eval = pd.concat([data.loc[good_eval], data.loc[bad_eval]])
eval
```

```python
# добавочные новости для train. пул новостей при активном обучении
indexes = news.index.tolist()

add_news = []

for index in indexes:
  if index not in good_news[:500] and index not in bad_news[:500]:
    add_news.append(index)
```

```python
add_news
```

#### Набор данных для обучения

```python
train['result_match'] = 0
eval['result_match'] = 0
```

```python
# разметка данных
indexes = train.index.tolist()

for index in indexes:
  if index in good_news[:500]:
    train.at[index, 'result_match'] = 1

indexes = eval.index.tolist()

for index in indexes:
  if index in good_news[:500]:
    eval.at[index, 'result_match'] = 1
```

```python
print(train['result_match'].unique())
print(train['result_match'].value_counts())
print(eval['result_match'].unique())
print(eval['result_match'].value_counts())
```

```python
data['result_match'] = 0
```

```python
final_train = pd.concat([train.sort_values(['match_id'], ascending = True), data.loc[add_news]])
final_train
```

```python
final_train.to_csv('train.csv', index=False, encoding='utf-8')
```
