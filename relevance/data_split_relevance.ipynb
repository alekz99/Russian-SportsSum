```python
import pandas as pd
import numpy as np
import random
from pprint import pprint
from IPython.display import clear_output 

data = pd.read_csv("/content/data.csv")
data
```

```python
news = data[['id', 'team1', 'team2', 'first_part_content', 'last_part_content', 'part_body']].drop_duplicates()
news
```

```python
good_news = []
indexes = news.index.tolist()
random.shuffle(indexes)

for index in indexes:
  print('len(good_news)=', len(good_news))
  if len(good_news) == 500:
    break
  row = news.loc[index]
  print(row['team1'], 'VS', row['team2'])
  pprint(row['first_part_content'], width=150)
  print('--------------------')
  pprint(row['last_part_content'], width=150)
  print('--------------------')
  pprint(row['part_body'], width=150)
  label = int(input())
  if label == 1:
    good_news.append(index)
  clear_output(True)
```

```python
good_news
```

```python
indexes = news.index.tolist()
random.shuffle(indexes)

bad_news = []

for index in indexes:
  if len(bad_news) == 500:
    break
  if index not in good_news:
    bad_news.append(index)
```

```python
bad_news
```

```python
good_train = good_news[:250]
good_eval = good_news[250:]

bad_train = bad_news[:250]
bad_eval = bad_news[250:]
```

```python
train = pd.concat([data.loc[good_train], data.loc[bad_train]])
train
```

```python
eval = pd.concat([data.loc[good_eval], data.loc[bad_eval]])
eval
```

```python
# добавочные новости для train. пул новостей при активном обучении
indexes = data.index.tolist()
random.shuffle(indexes)

add_news = []

for index in indexes:
  if len(add_news) == 9_500:
    break
  if index not in good_news and index not in bad_news:
    add_news.append(index)
```

```python
add_news
```

#### Набор данных для обучения

```python
train['relevance'] = 0
```

```python
# разметка данных
indexes = train.index.tolist()

for cnt, index in enumerate(indexes):
  print('№', cnt)
  row = train.loc[index]
  print(row['team1'], 'VS', row['team2'])
  pprint(row['first_part_content'], width=150)
  print('--------------------')
  pprint(row['last_part_content'], width=150)
  print('--------------------')
  pprint(row['part_body'], width=150)
  label = int(input())
  train.at[index, 'relevance'] = label
  clear_output(True)
```

```python
train['relevance'].unique()
```

```python
data['relevance'] = 0
```

```python
final_train = pd.concat([train.sort_values(['match_id'], ascending = True), data.loc[add_news]])
final_train
```

```python
final_train.to_csv('train.csv', index=False, encoding='utf-8')
```
